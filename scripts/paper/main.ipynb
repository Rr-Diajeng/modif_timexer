```python
import numpy as np
import pandas as pd

folder_path = '/home/citi/Documents/rr/programming/timexer/modif_timexer/scripts/paper/results/long_term_forecast_TimeXer_144_6_TimeXer_custom_ftMS_sl144_ll48_pl6_dm512_nh8_el3_dl1_df512_expand2_dc4_fc1_ebtimeF_dtTrue_Timexer-MS_0/'

pred_real = np.load(folder_path + 'pred_real.npy')   # (N, pred_len, n_features)
true_real = np.load(folder_path + 'true_real.npy')

# ── Flatten (N, pred_len, features) → (N*pred_len, features) ─────────────────
n, pred_len, n_features = pred_real.shape

pred_flat = pred_real[:, :, -1].flatten()   # target col only (OT)
true_flat = true_real[:, :, -1].flatten()

# ── Build DataFrame ───────────────────────────────────────────────────────────
df_result = pd.DataFrame({
    'true_real': true_flat,
    'pred_real': pred_flat,
    'error':     true_flat - pred_flat,
    'abs_error': np.abs(true_flat - pred_flat),
})

df_result.to_csv('prediction_result.csv', index=False)
print(f"Saved to {folder_path}prediction_result.csv")
print(df_result.head(10))
```
